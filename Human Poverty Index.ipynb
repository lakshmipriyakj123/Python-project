```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
Human_Poverty_Index=pd.read_csv('Human_Poverty_Index.csv')
plt.figure(figsize=(10,6))
sns.lineplot(data=Human_Poverty_Index,x='Country',y='Ranking',marker='o')
plt.title('Country vs Poverty Ranking')
plt.xlabel('Country(cm)')
plt.ylabel('Ranking(cm)')
plt.grid(True)
plt.show()
```

```python
Human_Poverty_Index=pd.read_csv('Human_Poverty_Index.csv')
plt.figure(figsize=(10,6))
sns.lineplot(data=Human_Poverty_Index,x='Country', y='Ranking', hue='HPI-2',marker='o')
plt.title('Country vs Poverty Ranking')
plt.xlabel('Country(cm)')
plt.ylabel('Ranking(cm)')
plt.grid(True)
plt.show()
```

```python
avg_People_lacking_functional_literacy_skills_percent = Human_Poverty_Index.groupby('Country')['Ranking'].mean().reset_index()
print(avg_People_lacking_functional_literacy_skills_percent)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=avg_People_lacking_functional_literacy_skills_percent,x='Country',y='Ranking',palette='Spectral')
plt.title('Average People lacking functional literacy skills percent ')
plt.xlabel('Country(cm)')
plt.ylabel('Average People lacking functional literacy skills percent(cm)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=Human_Poverty_Index,x='Country',y='Ranking',hue='HPI-2',palette='Set1')
plt.title('Country vs Ranking')
plt.xlabel('Country(cm)')
plt.ylabel('Ranking(cm)')
plt.legend(title='Poverty ratio')
plt.show()
```
